```python
import pandas as pd
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
data=pd.read_csv('Market_Basket_Optimisation.csv',header=None)
data.head()
```

# <u>1. Data exploration</u>

```python
data.shape
```

<b>From the previous output, each row represents the items bought by each (one) customer. The dataframe will be converted to a list of transactions, as the apriori algorithm required.</b>

```python
transactions=[]
for i in range(0,7501):
    transactions.append([str(data.values[i,j]) for j in range(0,20)])
```

```python
print(transactions[:2])
```

<b>After that, the list will be converted into a one-hot encoded DataFrame.</b>

```python
te=TransactionEncoder()
te_ary=te.fit(transactions).transform(transactions)
df=pd.DataFrame(te_ary,columns=te.columns_)
df.head()
```

# <u>2. Apriori model application</u>

<b>Let's select itemsets within a minimum support of 0.001%</b>

```python

itemsets=apriori(df,min_support=0.001,use_colnames=True)
itemsets.head(20)
```

<b>Next, let's train the apriori model with the rule which sorts the result by 'lift' as metric. </b>

```python
rules=association_rules(itemsets,metric='lift',min_threshold=0.5)
rules.head()
```
